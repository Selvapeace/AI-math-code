```python
import math
import numpy as np
```

```python
# load data
x = np.loadtxt('q1x.txt')
y = np.loadtxt('q1y.txt')
y = 2 * (y - 0.5)
```

```python
# set initial parameters
m = np.size(y)
n = (np.shape(x))[1]
w = np.mat(np.random.rand(n,1))
kesai = np.mat(np.random.rand(m,1))
b = np.mat(np.random.rand(1,1))
C = 1
```

```python
# transform the initial problem into a QP problem —— objective fuction
k = m + n + 1
z = np.vstack((w,kesai,b))
P = np.mat(np.zeros((k,k)))
for i in range(k):
    for j in range(k):
        if i == j and i < n:
            P[i,j] = 1
        else:
            continue
a = np.vstack((np.mat(np.zeros((n,1))),np.mat(np.ones((m,1))),np.mat(np.zeros((1,1)))))
q = C * a
# now the objective function:  f(z) = (1/2)* (z.T) * P * z + (q.T) * z
```

```python
# transform the initial problem into a QP problem —— constraint condition
g1 = np.mat(np.diag(y)) * x
g2 = np.mat(np.eye(m,m))
g3 = (np.mat(y)).T
G1 = np.hstack((g1,g2,g3))
g4 = np.mat(np.zeros((m,n)))
g5 = np.mat(np.eye(m,m))
g6 = np.mat(np.zeros((m,1)))
G2 = np.hstack((g4,g5,g6))
G = - np.vstack((G1,G2))
h = - np.vstack((np.mat(np.ones((m,1))),np.mat(np.zeros((m,1)))))
# now the constraint condition:  Gz <= h
```

```python
# solve the QP problem
from cvxopt import solvers, matrix 
P = matrix(P)
q = matrix(q)
G = matrix(G)
h = matrix(h)
sol = solvers.qp(P,q,G,h)
print(sol)
```

```python
# the optimal parameters
w_opt = (sol['x'])[:2]
b_opt = (sol['x'])[101]
```

```python
# visualize
import matplotlib.pyplot as plt
t = []
for i in range(99):
    if y[i] == 1:
        t.append('red')
    else:
        t.append('green')
plt.figure()
plt.scatter(list(x[:,0]),list(x[:,1]),s = 75,c = t,alpha = 0.5)
xp = np.linspace(min(x[:,0]),max(x[:,0]),100)
yp = - (w_opt[0] * xp + b_opt)/ w_opt[1]     # yp is the best segmentation hyperplane
y1 = - (w_opt[0] * xp + b_opt - 1)/ w_opt[1]     # y1 is the support vector for y = 1
y0 = - (w_opt[0] * xp + b_opt + 1)/ w_opt[1]     # y0 is the support vector for y = 0
plt.plot(xp,yp)
plt.plot(xp,y1,linestyle = '--',color = 'red')
plt.plot(xp,y0,linestyle = '--',color = 'green')
plt.title('decision boundary for a linear SVM classifier with C=1')
```
